```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
sns.set(style="ticks")
sns.set_palette("husl")

kfactors_full = pd.read_csv('kfactors.cxv')
kfactors_monoZ = kfactors_full.query('Model == "MonoZ_V1"')
kfactors_monoW = kfactors_full.query('Model == "MonoW_V1"')

kfactors = kfactors_monoZ.pivot("Dark Matter mass [GeV]", "Mediator mass [GeV]", "kFactor")

# Draw a heatmap with the numeric values in each cell
sns.heatmap(kfactors, annot=True, linewidths=.5, fmt='g', cmap="YlGnBu").invert_yaxis()
```

```python
kfactors = kfactors_monoW.pivot("Dark Matter mass [GeV]", "Mediator mass [GeV]", "kFactor")

# Draw a heatmap with the numeric values in each cell
sns.heatmap(kfactors, annot=True, linewidths=.5, fmt='g', cmap="YlGnBu").invert_yaxis()
```
