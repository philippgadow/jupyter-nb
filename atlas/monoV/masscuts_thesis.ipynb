# Mass Windows for the mono-W/Z tagger

## Z tagger

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
pt = np.arange(200.0, 2700.0)
```

```python
def lowercut_z_mass(pt):
    #if pt > 2500:
    #    return 69.4240299485
    return np.sqrt(np.power((-8911.32152025)/pt-(-72.3863305346),2)+np.power((0.0232216747865)*pt+(-67.1786329156),2))
def uppercut_z_mass(pt):
    #if pt > 2500:
    #    return 109.831901054
    return np.sqrt(np.power((428.215902774)/pt-(-105.538285882),2)+np.power((0.0192931557406)*pt+(-18.4246216168),2))

def uppercut_z_d2(pt):
    # if pt > 2500:
    #    return 1.90713763304
    return (0.86245030986)+(0.0010301169875)*pt+(-6.37151804335e-07)*np.power(pt,2.0)+(2.95602462111e-10)*np.power(pt,3.0)+(-5.54801878773e-14)*np.power(pt,4.0)
```

```python
print("mass cut for pT", lowercut_z_mass(534), uppercut_z_mass(534))
```

```python
f, axarr = plt.subplots(3, sharex=True, figsize=(8,8))
axarr[0].text(0.05, 0.75, '$Z$-Tagger: Massenfenster',transform=axarr[0].transAxes, fontsize=24)
axarr[0].set_ylabel('$m_{up}$ cut [GeV]')
axarr[0].plot(pt, uppercut_z_mass(pt), 'r--')

axarr[1].set_ylabel('$m_{low}$ cut [GeV]')
axarr[1].plot(pt, lowercut_z_mass(pt), 'b--')

axarr[2].set_xlabel('pt [GeV]')
axarr[2].set_ylabel('D2$_{up}$ cut')
axarr[2].text(0.4, 0.25, '$Z$-Tagger: D2-Schnitt',transform=axarr[2].transAxes, fontsize=24)
axarr[2].plot(pt, uppercut_z_d2(pt), 'g--')
plt.show()
```

### W tagger

```python
def lowercut_w_mass(pt):
    #if pt > 2500:
    #    return  60.0593267161
    return np.sqrt(np.power((-8680.2717943)/pt-(-54.2916959506),2)+np.power((0.0142530985867)*pt+(-67.640437403),2))
def uppercut_w_mass(pt):
    #if pt > 2500:
    #    return 99.824544933
    return np.sqrt(np.power((79.4993903251)/pt-(-94.0432300643),2)+np.power((0.0201350277197)*pt+(-16.9485211641),2))
def uppercut_w_d2(pt):
    # if pt > 2500:
    #    return 2.01439576478
    return (0.827098899953)+(0.00117390855197)*pt+(-7.27657125295e-07)*np.power(pt,2.0)+(3.46942900597e-10)*np.power(pt,3.0)+(-6.7087367778e-14)*np.power(pt,4.0)
```

```python
f, axarr = plt.subplots(3, sharex=True, figsize=(8,8))
axarr[0].text(0.05, 0.75, '$W$-Tagger: Massenfenster',transform=axarr[0].transAxes, fontsize=24)
axarr[0].set_ylabel('$m_{up}$ cut [GeV]')
axarr[0].plot(pt, uppercut_w_mass(pt), 'r--')

axarr[1].set_ylabel('$m_{low}$ cut [GeV]')
axarr[1].plot(pt, lowercut_w_mass(pt), 'b--')

axarr[2].set_xlabel('pt [GeV]')
axarr[2].set_ylabel('D2$_{up}$ cut')
axarr[2].text(0.4, 0.25, '$W$-Tagger: D2-Schnitt',transform=axarr[2].transAxes, fontsize=24)
axarr[2].plot(pt, uppercut_w_d2(pt), 'g--')
plt.show()
```
