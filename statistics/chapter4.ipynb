heuristic proof of the law of large numbers

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
# generate uniform distributions
d1 = np.random.uniform(0, 1, 5000)
d2 = np.random.uniform(0, 1, 5000)
d3 = np.random.uniform(0, 1, 5000)
```

```python
plt.hist(d1)
plt.show()
```

```python
plt.hist(d2)
plt.show()
```

```python
plt.hist(d3)
plt.show()
```

```python
# adding together independent distributions like those uniformly distributions above leads to a gaussian distribution
d4 = d1 + d2
d5 = d1 + d2 + d3
```

```python
plt.hist(d4)
plt.show()
```

```python
# to demonstrate this fact plot a Gaussian distribution on top with parameters mu = mu(d1) + mu(d2) + mu(d3) and sigma 0 sqrt(v1 + v2 + v3)
plt.hist(d5, normed=True)
mu = 1.5
sigma = np.sqrt(3 * 1. / 12.)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))
plt.show()
```
