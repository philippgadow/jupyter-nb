```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bkg(x, gamma):
    return np.exp(-x/gamma)/gamma

def sig(x, mu, sigma, C):
    return C * np.exp(-(x-mu)**2/(2*sigma**2))/(sigma*np.sqrt(2 * np.pi))

def sigbkg(x,gamma,mu,sigma,C):
    return sig(x,mu,sigma,C) + np.exp(-x/gamma)/gamma

def y(x, gamma, mu, sigma, C):
    return sig(x, mu, sigma, C) / bkg(x, gamma)
```

```python
x = np.arange(2., 40., 0.2)
plt.plot(x, bkg(x, 10), 'r--', sig(x,7,0.5,0.08), 'b--')
plt.xlim([0,40])
plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```

```python
x = np.random.uniform(2., 40., 300)
plt.plot(x,-2*np.log(y(x,10,5,0.5,0.08)), "o")
plt.show()
```

```python
def Pois(n, mu, s, b):
    return (mu*s + b)**n / np.factorial(n) * np.exp(-(mu*s+b))
def L(mu, n, s, b):
    Pois(n, mu, s, b)
```
