## pyhf example
taken from https://diana-hep.org/pyhf/examples.html
works only with python2 since my python3 version is 3.5 and that is not supported by the latest pyhf install

```python
import pyhf
import numpy as np
```

```python
def doCLS(sig):
    pdf = pyhf.simplemodels.hepdata_like(signal_data=sig, bkg_data=[21700, 14400, 1120, 130], bkg_uncerts=[140, 110, 25, 7.2])
    CLs_obs, CLs_exp = pyhf.utils.hypotest(1.0, [21818, 14350, 1128., 119] + pdf.config.auxdata, pdf, return_expected=True)
    return CLs_obs, CLs_exp
```

```python
def checkSignal(xsec, eff500, eff150, eff200, eff350):
    sig = xsec * 79.8 * np.array([eff150, eff200, eff350, eff500])
    CLs_obs, CLs_exp = doCLS(sig.tolist())
    print('Observed: {}, Expected: {}, Theory: {}, Excluded: {}'.format(CLs_obs, CLs_exp, xsec, (CLs_obs < xsec)))

checkSignal(3.4079035334678514,0.08379121190471013,0.03712378329919567,0.0940846942190662,0.04882149421035165)
```
