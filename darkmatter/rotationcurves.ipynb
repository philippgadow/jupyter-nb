## Galactic rotation curves and Dark Matter

This notebook is designed to play around with various models for galactic dynamics and dark matter models to get an understanding for the motivation of Dark Matter in theories to explain galactic rotation curves.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from galpy.potential import MiyamotoNagaiPotential, HernquistPotential, NFWPotential
from galpy.potential import MWPotential2014
from galpy.potential import plotRotcurve
```

```python
mp = MiyamotoNagaiPotential(a=0.5, b=0.0375, normalize=.6)
np = NFWPotential(a=4.5, normalize=.35)
hp = HernquistPotential(a=0.6 / 8, normalize=0.05)
```

```python
mp.plotRotcurve(Rrange=[0.01, 10.], grid=1001, overplot=True)
hp.plotRotcurve(Rrange=[0.01, 10.], grid=1001, overplot=True)
np.plotRotcurve(Rrange=[0.01, 10.], grid=1001, overplot=True)

plotRotcurve([mp, hp], Rrange=[0.01, 10.], grid=1001, overplot=True)
plotRotcurve([mp, hp, np], Rrange=[0.01, 10.], grid=1001, overplot=True)
```
